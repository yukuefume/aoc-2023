```python
from aocd import get_data, submit

data = get_data(day=3, year=2023)
data = data.split('\n')
data
```

```python
import re

gear_ratio_sum = 0

def num_at_pos(line_number, pos):
  query = r'[0-9]+'
  result_iter = re.finditer(query, data[line_number])

  for result in result_iter:
    if pos >= result.start() and pos < result.end():
      number = result.group()

      try:
        return (int(number), (line_number, result.start(), result.end()))
      except Exception:
        return None

def extract_part_no(pos, line_number, line, size):
  part_number = {}

  # Boundry checks
  left_pos = pos-1 if (pos-1) >= 0 else -1
  right_pos = pos+1 if (pos+1) < len(line) else -1
  up_pos = line_number-1 if (line_number-1) >= 0 else -1
  down_pos = line_number+1 if (line_number+1) < size else -1

  l = 0
  r = 0
  u = 0
  d = 0
  ul = 0
  ur = 0
  dl = 0
  dr = 0

  if left_pos >= 0:
    l = (line_number, pos-1)
    num = num_at_pos(*l)
    if num:
      v, k =num
      part_number[k] = v

  if right_pos >= 0:
    r = (line_number, pos+1)
    num = num_at_pos(*r)
    if num:
      v, k =num
      part_number[k] = v

  if up_pos >= 0:
    u = (line_number-1, pos)
    num = num_at_pos(*u)
    if num:
      v, k =num
      part_number[k] = v

  if down_pos >= 0:
    d = (line_number+1, pos)
    num = num_at_pos(*d)
    if num:
      v, k =num
      part_number[k] = v

  if up_pos >= 0 and left_pos >= 0:
    ul = (line_number-1, pos-1)
    num = num_at_pos(*ul)
    if num:
      v, k =num
      part_number[k] = v

  if up_pos >= 0 and right_pos >= 0:
    ur = (line_number-1, pos+1)
    num = num_at_pos(*ur)
    if num:
      v, k =num
      part_number[k] = v

  if down_pos >= 0 and left_pos >= 0:
    dl = (line_number+1, pos-1)
    num = num_at_pos(*dl)
    if num:
      v, k =num
      part_number[k] = v

  if down_pos >= 0 and right_pos >= 0:
    dr = (line_number+1, pos+1)
    num = num_at_pos(*dr)
    if num:
      v, k =num
      part_number[k] = v

  print(f'parts: {part_number}')
  if len(part_number) == 2:
    print('Its a valid part')

    gear_ratio = 1
    for _, p in part_number.items():
      gear_ratio = gear_ratio * p
    print(gear_ratio)
    return gear_ratio
  return 0

def symbol_pos(line):
  query = r'\*'  # Find all '*'
  result_iter = re.finditer(query, line)

  return [symbol.start() for symbol in result_iter]

```

```python
for index, line in enumerate(data):
  symbol_postions = symbol_pos(line)
  print(f'Line: {index}: symbols found: {len(symbol_postions)}')
  print(gear_ratio_sum)

  for pos in symbol_postions:
    print(f'Symbol at pos: {pos}')
    gear_ratio = extract_part_no(pos, index, line, len(data))
    gear_ratio_sum = gear_ratio_sum + gear_ratio

  print('')

print(gear_ratio_sum)
```
