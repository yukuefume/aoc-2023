```python
sample_data = """seeds: 79 14 55 13

seed-to-soil map:
50 98 2
52 50 48

soil-to-fertilizer map:
0 15 37
37 52 2
39 0 15

fertilizer-to-water map:
49 53 8
0 11 42
42 0 7
57 7 4

water-to-light map:
88 18 7
18 25 70

light-to-temperature map:
45 77 23
81 45 19
68 64 13

temperature-to-humidity map:
0 69 1
1 0 69

humidity-to-location map:
60 56 37
56 93 4"""

sample_data = sample_data.split('\n')
sample_data
```

```python
input_data = sample_data

seeds = [int(n) for n in input_data[0].split(':')[1].strip().split()]
seeds
```

```python
import re

mappings = []
map_input = input_data[2::]

value = []
for i, m in enumerate(map_input):
  # Check if contains a number
  if bool(re.search(r'\d', m)):
    value.append([int(n) for n in m.split()])
  elif m == '':
    mappings.append(value)
    value = []  # Reset value array

# Add last value
if value:
  mappings.append(value)

mappings
```

```python
def next_value(value_list, num):
  for source, dest, v_range in value_list:
    # Within range
    if dest <= num < (dest+v_range):
      # Calculate difference
      diff = (dest-source) * -1
      return num + diff
  return num
```

```python
location = []

for seed in seeds:
  num = seed
  for category in mappings:
    num = next_value(category, num)
  location.append(num)

location
```

```python
low_location = min(location)
low_location
```
